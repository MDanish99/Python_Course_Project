I scaled the data first in order to perform dimensional reduction. I first found the optimum number of clusters using the fact that inertia should be low. I choose number of clusters to be 7 as the grpah shows that as number of clusters are 7 or above, the rate of fall of inertia becomes stable.

With help of PCA, I tried to visualise the clusters trying to reduce data from 29 dimensions to 2 dimension. Then I tried to identy best number of components that cover 95% of the total variance in the components.

I came to a conclusion that using 20 components will cover approximately 95% of the variance. Using PCA with 20 components, we can see that inertia decreased and rate of fall was also stable in comparison to previously used PCA with 2 components.

With reference to the results:

Random Forrest Classifier gives the highest weighted average precision score which means it has the highest ratio of true positives to all positives, although SVC and Gaussian Process Classifier have similar precision scores. It means that if the model predicts that a transaction is Fraud, it is correct 95% of the time.

Random Forrest Classifer has also highest weighted average recall score which means that out of all the fraud transactions, it is able to identify 95% of them corrcectly. Gaussian Process have a slightly lower recall score while SVC have the lowest recall scores out of these three models.

There is always a trade-off between precision and recall score, that's why f1 score is also important as it provides harmonic mean of precision and recall. Random Forrest Classifier also has the highest f1 score due to which it again becomes top preference among all three. Moreover, ROC_AUC scores that measures that how well model performes for distinguishing transactions between fraudulent and non-fraud, Random Forrest Classifier again performs the best among the three models.

Therefore, my conclusion is that Random Forrest Classifer is the best performing models among the above mentioned three classification models as it has the highest precision, recall, f1 and ROC_AUC scores. Gaussian Process Classifier also have performed very well and have very similar scores, while Support Vector Machines didn't performed as well as the other two models.
